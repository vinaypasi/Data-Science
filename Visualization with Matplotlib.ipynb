```python
plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); 
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
 # For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':'); # dotted
```

```python
plt.plot(x, x + 0, '-g') # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r'); # dotted red
```

```python
# Visualizing Errors

# Basic errorbars

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0,10,50)
dy = 0.8
y = np.sin(x) + dy*np.random.randn(50)

plt.errorbar(x,y,yerr=dy,fmt='.k')
```

```python
plt.errorbar(x,y,yerr=dy,fmt='o',color='black' , ecolor='lightgray' , elinewidth=3, capsize=0)
```

```python
# Continuous Errors

from sklearn.gaussian_process import GaussianProcessRegressor

model = lambda x: x * np.sin(x)
xdata = np.array([1,3,5,6,8])
ydata = model(xdata)

gp = GaussianProcessRegressor( theta0=1e-2 , thetaL = 1e-4 , thetU = 1E-1 , random_start = 100)
gp.fit(xdata[:,np.newaxis],ydata)

xfit = np.linspace(0,10,1000)
yfit, MSE = gp.predict(xfit[:,np.newaxis] , eval_MSE = True)
dyfit = 2*np.sqrt(MSE)
```

```python
# Density and contour plots

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
# Visualizing a three-Dimensional Data

def f(x,y):
    return np.sin(x)**10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)

X, Y = np.meshgrid(x,y)
Z = f(X,Y)

print(X.shape,Y.shape,Z.shape)
```

```python
plt.contour(X,Y,Z, colors='black' );
```

```python
plt.contour(X,Y,Z,20, cmap='RdGy');
```

```python
plt.contourf(X,Y,Z,20, cmap='RdGy')
plt.colorbar();
```

```python
plt.imshow(Z , extent=[0,5,0,5] , origin='lower' , cmap='RdGy' )
plt.colorbar()
plt.axis('image');
```

```python
contours = plt.contour(X,Y,Z,3, colors='black')
plt.clabel(contours, inline=True , fontsize=8)
plt.imshow(Z , extent=[0,5,0,5] , origin='lower' , cmap='RdGy' , alpha = 0.5)
plt.colorbar()
```

```python
# Histogram, Binnning,  and Density

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np

data = np.random.randn(1000)
```

```python
plt.hist(data,edgecolor='black');
```

```python
plt.hist(data, bins = 30 , normed=True , alpha=0.5 , histtype='stepfilled' , color='steelblue' , edgecolor='none' )
```

```python
x1 = np.random.normal(0,0.8,1000)
x2 = np.random.normal(-2,1,1000)
x3 = np.random.normal(3,2,1000)

kwargs = dict(histtype='stepfilled' , alpha = 0.2 , bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

```python
counts, bin_edges = np.histogram(data,bins=5)
print(counts,bin_edges)
```

```python
mean = [0,0]
cov = [[1,1],[1,2]]
x,y = np.random.multivariate_normal(mean,cov,size=10000).T
```

```python
plt.hist2d(x,y,bins=30,cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
mean = [0,0]
cov = [[4,4],[4,5]]
x,y = np.random.multivariate_normal(mean,cov,size=10000).T
```

```python
plt.hist2d(x,y,bins=30,cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
counts,x_edges,y_edges = np.histogram2d(x,y,bins=30)
print(counts,x_edges,y_edges)
```
